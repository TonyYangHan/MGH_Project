### Combine dataframes to get gene names

```r
base <- "result/DEG_Bing_Yang/"
files <- list.files(base, pattern="\\.tsv$", recursive=TRUE, full.names=TRUE)
files <- files[grepl("lower", files, ignore.case=TRUE)]

rows <- unique(unlist(lapply(files, function(f) {
  x <- tryCatch(fread(f, sep="\t", header=TRUE, select=1),
                error=function(e) fread(f, sep="\t", header=FALSE, select=1))
  unlist(x[[1]])
})))

fwrite(data.table(rows), paste0(base, "down_genes.txt"), col.names=FALSE)
```
